```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import pandas as pd 
```

```python
DataSet = pd.read_csv('Data/CleanData.csv')
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(DataSet['year'], bins=30, kde=True, color="blue")
plt.title("Distribution of year")
plt.xlabel("year")
plt.ylabel("Frequence")
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(DataSet['price'], bins=30, kde=True, color="blue")
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Frequence")
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(6, 4))
sns.histplot(DataSet['puissance'], bins=30, kde=True, color="blue")
plt.title("Distribution of Puissance")
plt.xlabel("Puissance")
plt.ylabel("Frequence")
plt.show()

```

```python
cat_df = DataSet.select_dtypes(include=['object']).drop('url'  ,axis=1)
```

```python
# plotting a histogram for each categorcal column.
for col in cat_df:
    plt.figure(figsize=(13, 4))
    cat_df[col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
numerical_column  = DataSet.select_dtypes(include=['int64' ,'float64'])
correlation_matrix = DataSet.corr(numeric_only=True)
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix  ,vmin=-1 , vmax=1 , annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.pairplot(DataSet, hue='type_boit', diag_kind='kde', 
             vars=['puissance', 'kilometrage' , 'price' , 'year'], 
             palette='Set1', markers=['o', 's', 'D', '^'])
plt.suptitle('Relations between variables by type_boit', y=1.02)
plt.show()
```

```python
sns.set_style("whitegrid")

sns.pairplot(DataSet, hue='type_carburant', diag_kind='kde', 
             vars=['puissance', 'kilometrage' , 'price' , 'year'], 
             palette='Set1', markers=['o', 's', 'D', '^'])
plt.suptitle('Relations between variables by type_carburant', y=1.02)
plt.show()
```

```python

sns.violinplot(x='type_carburant', y='price', hue='type_boit', split=True, data=DataSet, 
               palette='muted', inner='quartile')
plt.title('distribution of price by type_carburant and type_boit')
plt.xlabel('type_carburant')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='puissance', y='price', data=DataSet, 
            scatter_kws={'alpha': 0.5, 'color': 'blue'}, 
            line_kws={'color': 'red', 'lw': 2})
plt.ylim(0, 450000)
plt.title('Relation between price and puissance')
plt.xlabel('Puissance (%)')
plt.ylabel('Price')
plt.show()
## display relation between price and puissance with a regression line
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='kilometrage', y='price', data=DataSet, 
            scatter_kws={'alpha': 0.5, 'color': 'blue'}, 
            line_kws={'color': 'red', 'lw': 2})
plt.ylim(0, 400000)
plt.title('Relation between price and kilometrage ')
plt.xlabel('kilometrage')
plt.ylabel('price')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.regplot(x='year', y='price', data=DataSet, 
            scatter_kws={'alpha': 0.5, 'color': 'blue'}, 
            line_kws={'color': 'red', 'lw': 2})
plt.ylim(0, 400000)
plt.title('Relation between price and year')
plt.xlabel('year')
plt.ylabel('price')
plt.show()
```
