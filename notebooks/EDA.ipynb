```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import pandas as pd 
```

```python
DataSet = pd.read_csv('Data/CleanData.csv')
```

```python
print(DataSet.shape)
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(DataSet['year'], bins=30, kde=True, color="blue")
plt.title("Distribution of year")
plt.xlabel("year")
plt.ylabel("Frequence")
plt.show()
```

```python


plt.figure(figsize=(8, 6))
plt.hist(DataSet['price'], bins=30, color='skyblue', edgecolor='black')

# Add lines for mean, median, mode
plt.axvline(DataSet['price'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(DataSet['price'].median(), color='green', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(DataSet['price'].mode().iloc[0], color='orange', linestyle='dashed', linewidth=2, label='Mode')

plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(6, 4))
sns.histplot(DataSet['puissance'], bins=30, kde=True, color="blue")
plt.title("Distribution of Puissance")
plt.xlabel("Puissance")
plt.ylabel("Frequence")
plt.show()

```

```python
plt.figure(figsize=(4, 4))
type_boit_distribution = DataSet['type_boit'].value_counts()
plt.pie(type_boit_distribution, labels=type_boit_distribution.index, autopct='%1.1f%%',
        explode=[0,0.2] , startangle=90, wedgeprops={'edgecolor':'k'}, colors=plt.cm.Paired.colors)
plt.title('Distribution of Car Fuel')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), title='type_boit', ncol=len(type_boit_distribution))
plt.show()
```

```python
plt.figure(figsize=(4, 4))
type_carburant_distribution = DataSet['type_carburant'].value_counts()
plt.pie(type_carburant_distribution, labels=type_carburant_distribution.index, autopct='%1.1f%%',
       startangle=90, wedgeprops={'edgecolor':'k'}, colors=plt.cm.Paired.colors)
plt.title('Distribution of Car Fuel')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), title='type_carburant', ncol=len(type_carburant_distribution))
plt.show()
```

```python
cat_df = DataSet.select_dtypes(include=['object']).drop(['url' ,'type_boit' , 'type_carburant']  ,axis=1)
```

```python
# plotting a histogram for each categorcal column.
for col in cat_df:
    plt.figure(figsize=(13, 4))
    top_20 = cat_df[col].value_counts().head(20)
    top_20.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
numerical_column  = DataSet.select_dtypes(include=['int64' ,'float64']).drop('index' ,axis=1)
correlation_matrix = DataSet.corr(numeric_only=True)
sns.heatmap(correlation_matrix  ,vmin=-1 , vmax=1 , annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.pairplot(DataSet, hue='type_boit', diag_kind='kde', 
             vars=['puissance', 'kilometrage' , 'price' , 'year'], 
             palette='Set1', markers=['o', 's', 'D', '^'])
plt.suptitle('Relations between variables by type_boit', y=1.02)
plt.show()
```

```python
sns.set_style("whitegrid")

sns.pairplot(DataSet, hue='type_carburant', diag_kind='kde', 
             vars=['puissance', 'kilometrage' , 'price' , 'year'], 
             palette='Set1', markers=['o', 's', 'D', '^'])
plt.suptitle('Relations between variables by type_carburant', y=1.02)
plt.show()
```

```python

sns.violinplot(x='type_carburant', y='price', hue='type_boit', split=True, data=DataSet, 
               palette='muted', inner='quartile')
plt.title('distribution of price by type_carburant and type_boit')
plt.xlabel('type_carburant')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='puissance', y='price', data=DataSet, 
            scatter_kws={'alpha': 0.5, 'color': 'teal'}, 
            line_kws={'color': 'red', 'lw': 2})
plt.title('Relation between price and puissance')
plt.xlabel('Puissance (%)')
plt.ylabel('Price')
plt.show()
## display relation between price and puissance with a regression line
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='kilometrage', y='price', data=DataSet, 
            scatter_kws={'alpha': 0.5, 'color': 'teal'}, 
            line_kws={'color': 'red', 'lw': 2})
plt.title('Relation between price and kilometrage ')
plt.xlabel('kilometrage')
plt.ylabel('price')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.regplot(x='year', y='price', data=DataSet, 
            scatter_kws={'alpha': 0.5, 'color': 'teal'}, 
            line_kws={'color': 'red', 'lw': 2})
plt.title('Relation between price and year')
plt.xlabel('year')
plt.ylabel('price')
plt.show()
```
